```python
import pandas as pd
import numpy as np
```

```python
# load data
train_data = pd.read_csv('train.csv', index_col = 'PassengerId')
```

```python
test_data = pd.read_csv('test.csv', index_col = 'PassengerId')
```

```python
# check data
train_data.head()
```

```python
# seperate target from predictors
y = train_data.Survived
```

```python
X = train_data.drop(['Survived'], axis=1)
```

```python
X.head()
```

```python
X_test_full = test_data.copy()
```

```python
from sklearn.model_selection import train_test_split
# break off validation set from training data
X_train_full, X_valid_full, y_train_full, y_valid_full = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=0)
```

```python
# Selecting low cardinality columns
low_cardinality_cols = [cname for cname in X_train_full.columns if X_train_full[cname].nunique() < 10 and
                       X_train_full[cname].dtype == 'object']
```

```python
# Selecting numeric columns
num_cols = [cname for cname in X_train_full.columns if X_train_full[cname].dtype in ['int64', 'float64']]
```

```python
# Keep selected columns only
my_cols = low_cardinality_cols + num_cols
X_train = X_train_full[my_cols].copy()
X_valid = X_valid_full[my_cols].copy()
X_test = X_test_full[my_cols].copy()
```

```python
# One_Hot_Encode the data
X_train = pd.get_dummies(X_train)
X_valid = pd.get_dummies(X_valid)
X_test = pd.get_dummies(X_test)
X_train, X_vaild = X_train.align(X_valid, join='left', axis=1)
X_train, X_test = X_train.align(X_test, join='left', axis=1)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
X_valid.head()
```

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error
```

```python
# Creating XGB model and fitting with data
my_model = XGBRegressor(n_estimators=100, learning_rate=0.05, n_jobs=2)

my_model.fit(X_train, y_train_full,
            early_stopping_rounds=5,
            eval_set=[(X_valid, y_valid_full)],
            verbose=False)
```

```python
# Predicting validation data
predictions = my_model.predict(X_valid)
```

```python
# Finding mean absolute error
mae = mean_absolute_error(predictions, y_valid_full)
```

```python
print(mae)
```

```python
preds = my_model.predict(X_test)
```

```python
output = pd.DataFrame({'PassengerId': X_test.index,
                      'Survived':preds})
```

```python
output.to_csv('submissionv1.csv', index=False)
```
